# Solar Power Generation and Plant Performance Analyzer

## Objective

The goal of this notebook is to prepare the raw solar power generation and weather datasets for analysis and machine learning.

The tasks performed in this notebook include:

- Loading the raw datasets
- Inspecting data quality
- Handling missing values and duplicates
- Converting data types
- Merging generation and weather datasets
- Saving a cleaned dataset for downstream analysis

The processed dataset generated in this notebook will be used throughout the rest of the project.

```python

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)
```

```python

plant1_gen = pd.read_csv("Plant_1_Generation_Data.csv")
plant1_weather = pd.read_csv("Plant_1_Weather_Sensor_Data.csv")
plant2_gen = pd.read_csv("Plant_2_Generation_Data.csv")
plant2_weather = pd.read_csv("Plant_2_Weather_Sensor_Data.csv")
```

```python
def dataset_overview(df, name):

    print("="*70)
    print(name)
    print("="*70)

    print(f"Shape : {df.shape}")

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")
    print(df.duplicated().sum())

    print("\nFirst Five Rows")
    display(df.head())
```

```python
dataset_overview(plant1_gen,"Plant 1 Generation")
dataset_overview(plant1_weather,"Plant 1 Weather")

dataset_overview(plant2_gen,"Plant 2 Generation")
dataset_overview(plant2_weather,"Plant 2 Weather")
```

```python
datasets = [
    plant1_gen,
    plant1_weather,
    plant2_gen,
    plant2_weather
]

for df in datasets:
    df["DATE_TIME"] = pd.to_datetime(df["DATE_TIME"])
```

```python
quality_report = pd.DataFrame()

for name, df in {
    "Plant1_Generation": plant1_gen,
    "Plant1_Weather": plant1_weather,
    "Plant2_Generation": plant2_gen,
    "Plant2_Weather": plant2_weather
}.items():

    quality_report.loc[name,"Rows"] = df.shape[0]
    quality_report.loc[name,"Columns"] = df.shape[1]
    quality_report.loc[name,"Missing"] = df.isnull().sum().sum()
    quality_report.loc[name,"Duplicates"] = df.duplicated().sum()
```

```python
quality_report
```

```python
# ============================================================
# Data Validation
# ============================================================

def validate_data(df, name):

    print("="*70)
    print(name)
    print("="*70)

    print("Negative DC_POWER:",
          (df["DC_POWER"] < 0).sum() if "DC_POWER" in df.columns else "N/A")

    print("Negative AC_POWER:",
          (df["AC_POWER"] < 0).sum() if "AC_POWER" in df.columns else "N/A")

    print("Negative DAILY_YIELD:",
          (df["DAILY_YIELD"] < 0).sum() if "DAILY_YIELD" in df.columns else "N/A")

    print("Negative IRRADIATION:",
          (df["IRRADIATION"] < 0).sum() if "IRRADIATION" in df.columns else "N/A")

    print("Ambient Temp < -20:",
          (df["AMBIENT_TEMPERATURE"] < -20).sum()
          if "AMBIENT_TEMPERATURE" in df.columns else "N/A")

    print("Module Temp < -20:",
          (df["MODULE_TEMPERATURE"] < -20).sum()
          if "MODULE_TEMPERATURE" in df.columns else "N/A")
```

```python
validate_data(plant1_gen, "Plant 1 Generation")
validate_data(plant2_gen, "Plant 2 Generation")

validate_data(plant1_weather, "Plant 1 Weather")
validate_data(plant2_weather, "Plant 2 Weather")
```

```python
def date_range(df,name):

    print("="*80)
    print(name)
    print("="*80)

    print("Start Date :",df["DATE_TIME"].min())
    print("End Date   :",df["DATE_TIME"].max())
    print("Number of Records :",len(df))
    print()


date_range(plant1_gen,"Plant 1 Generation")
date_range(plant1_weather,"Plant 1 Weather")

date_range(plant2_gen,"Plant 2 Generation")
date_range(plant2_weather,"Plant 2 Weather")
```

```python
print("="*80)
print("Plant Summary")
print("="*80)

print("\nPlant 1")
print("Plant ID :",plant1_gen["PLANT_ID"].unique()[0])
print("Generation Inverters :",plant1_gen["SOURCE_KEY"].nunique())
print("Weather Sensors :",plant1_weather["SOURCE_KEY"].nunique())

print("\nPlant 2")
print("Plant ID :",plant2_gen["PLANT_ID"].unique()[0])
print("Generation Inverters :",plant2_gen["SOURCE_KEY"].nunique())
print("Weather Sensors :",plant2_weather["SOURCE_KEY"].nunique())
```

```python
# =============================================================================
# 10. Merge Plant 1
# =============================================================================

plant1 = pd.merge(
    plant1_gen,
    plant1_weather.drop(columns=["SOURCE_KEY"]),
    on=["DATE_TIME","PLANT_ID"],
    how="left"
)

plant1["PLANT_NAME"] = "Plant_1"


# =============================================================================
# 11. Merge Plant 2
# =============================================================================

plant2 = pd.merge(
    plant2_gen,
    plant2_weather.drop(columns=["SOURCE_KEY"]),
    on=["DATE_TIME","PLANT_ID"],
    how="left"
)

plant2["PLANT_NAME"] = "Plant_2"


# =============================================================================
# 12. Combine Both Plants
# =============================================================================

solar_complete = pd.concat(
    [plant1,plant2],
    ignore_index=True
)

print("="*80)
print("Merged Dataset")
print("="*80)

print("Shape :",solar_complete.shape)

display(solar_complete.head())
```

```python
# =============================================================================
# 13. Final Validation
# =============================================================================

print("=" * 80)
print("FINAL DATASET VALIDATION")
print("=" * 80)

print("\nDataset Information\n")
solar_complete.info()

print("\n" + "=" * 80)
print("Missing Values")
print("=" * 80)
print(solar_complete.isnull().sum())

print("\nTotal Missing Values :", solar_complete.isnull().sum().sum())

print("\n" + "=" * 80)
print("Duplicate Records")
print("=" * 80)
print("Duplicate Rows :", solar_complete.duplicated().sum())

print("\n" + "=" * 80)
print("Preview of Processed Dataset")
print("=" * 80)
display(solar_complete.head())


# =============================================================================
# 14. Save Processed Dataset
# =============================================================================

import os

# Create processed folder if it does not exist
os.makedirs("../data/processed", exist_ok=True)

plant1_path = "../data/processed/plant1_clean.csv"
plant2_path = "../data/processed/plant2_clean.csv"
solar_path = "../data/processed/solar_complete.csv"

plant1.to_csv(plant1_path, index=False)
plant2.to_csv(plant2_path, index=False)
solar_complete.to_csv(solar_path, index=False)

print("\nDatasets saved successfully!")

print("\nSaved Files")
print("-" * 40)
print(f"Plant 1 Dataset : {plant1_path}")
print(f"Plant 2 Dataset : {plant2_path}")
print(f"Combined Dataset: {solar_path}")


# =============================================================================
# 15. Verify Saved Dataset
# =============================================================================

print("\n" + "=" * 80)
print("VERIFY SAVED DATASET")
print("=" * 80)

saved_dataset = pd.read_csv(solar_path)

print(f"Saved Dataset Shape : {saved_dataset.shape}")

if saved_dataset.shape == solar_complete.shape:
    print("Dataset verification successful.")
else:
    print("Warning: Saved dataset shape does not match.")


# =============================================================================
# 16. Notebook Summary
# =============================================================================

print("\n" + "=" * 80)
print("NOTEBOOK 01 SUMMARY")
print("=" * 80)

summary = pd.DataFrame({
    "Metric": [
        "Raw Datasets Loaded",
        "Generation Datasets",
        "Weather Datasets",
        "Merged Dataset Created",
        "Final Dataset Rows",
        "Final Dataset Columns",
        "Missing Values",
        "Duplicate Rows"
    ],
    "Value": [
        4,
        2,
        2,
        "Yes",
        solar_complete.shape[0],
        solar_complete.shape[1],
        solar_complete.isnull().sum().sum(),
        solar_complete.duplicated().sum()
    ]
})

display(summary)

print("\nOutput Files")
print("-" * 40)
print("✓ plant1_clean.csv")
print("✓ plant2_clean.csv")
print("✓ solar_complete.csv")

print("\nNext Notebook")
print("-" * 40)
print("Notebook 02 : Exploratory Data Analysis (EDA)")

print("\nNotebook 01 completed successfully!")
```

```python
# =============================================================================
# Handle Missing Values After Merge
# =============================================================================

print("=" * 80)
print("Handling Missing Values")
print("=" * 80)

print(f"Rows Before Cleaning : {solar_complete.shape[0]}")

missing_rows = solar_complete.isnull().any(axis=1).sum()

print(f"Rows with Missing Values : {missing_rows}")

# Remove rows with missing weather information
solar_complete = solar_complete.dropna().reset_index(drop=True)

print(f"Rows After Cleaning : {solar_complete.shape[0]}")

print(f"Remaining Missing Values : {solar_complete.isnull().sum().sum()}")


# =============================================================================
# Final Integrity Check
# =============================================================================

assert solar_complete.isnull().sum().sum() == 0
assert solar_complete.duplicated().sum() == 0

print("All integrity checks passed.")
```
